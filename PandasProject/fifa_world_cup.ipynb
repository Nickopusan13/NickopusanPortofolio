```python
import pandas as pd
```

```python
match_df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\Fifa World Cup\World Cup Matches Raw.csv")
rank_df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\Fifa World Cup\Fifa Ranking 2022.csv")
point_df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\Fifa World Cup\Fifa Host.csv")
```

```python
match_df
```

```python
match_df.drop(["home_yellow_card_long", "away_yellow_red_card", "home_yellow_red_card", "away_red_card", "home_red_card", "home_substitute_in_long", "away_substitute_in_long", "home_penalty_shootout_goal_long", "away_penalty_shootout_goal_long", "home_penalty_shootout_miss_long", "away_penalty_shootout_miss_long"], axis=1, inplace=True)
```

```python
match_df.drop(["away_yellow_card_long", "away_penalty_miss_long", "home_penalty_miss_long"], axis=1, inplace=True)
```

```python
match_df.drop(["home_xg", "away_xg"], axis=1, inplace=True)
```

```python
match_df.info()
```

```python
match_df['Date'] = pd.to_datetime(match_df['Date'])
match_df['Date'] = match_df['Date'].dt.date
```

```python
match_df[['Referee', 'AR1', 'AR2','4th', 'VAR']] = match_df['Officials'].str.split('·', expand=True)
match_df
```

```python
match_df.drop(columns='Officials', inplace=True)
```

```python
match_df['home_goal'] = match_df.apply(lambda row: f"{row['home_goal']} | {row['home_penalty_goal']}" if pd.notna(row['home_penalty_goal']) else row['home_goal'], axis=1)
match_df['away_goal'] = match_df.apply(lambda row: f"{row['away_goal']}|{row['away_penalty_goal']}" if pd.notna(row['away_penalty_goal']) else row['away_goal'], axis=1)
```

```python
match_df.drop(columns=['home_penalty_goal', 'away_penalty_goal', 'Notes', 'home_goal_long', 'away_goal_long'], inplace=True)
```

```python
match_df['home_goal'] = match_df.apply(lambda row: f"{row['home_goal']}|{row['home_own_goal']}" if pd.notna(row['home_own_goal']) else row['home_goal'], axis=1)
match_df['away_goal'] = match_df.apply(lambda row: f"{row['away_goal']}|{row['away_own_goal']}" if pd.notna(row['away_own_goal']) else row['away_goal'], axis=1)
```

```python
match_df.drop(columns=['away_own_goal', 'home_own_goal'], inplace=True)
```

```python
match_df['home_penalty'] = match_df['home_penalty'].astype('Int64')
match_df['away_penalty'] = match_df['away_penalty'].astype('Int64')
```
