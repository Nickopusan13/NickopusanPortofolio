```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\mochw\Downloads\unemployment-rate.filtered\unemployment-rate.csv")
df2 = pd.read_csv(r"C:\Users\mochw\Downloads\country-and-continent-codes-list-csv.csv")
```

```python
df.dropna(inplace=True)
df = df.drop(columns= ['time', 'Year'])
df.rename(columns={'Unemployment, total (% of total labor force) (modeled ILO estimate)': 'Unemployment By %'}, inplace=True)
df.rename(columns={'Entity': 'Country'}, inplace=True)
df = pd.merge(df, df2[['Three_Letter_Country_Code', 'Continent_Name']], left_on='Code',  right_on='Three_Letter_Country_Code', how='left')
df.drop(columns='Three_Letter_Country_Code', inplace=True)
```

```python
df.rename(columns={'Continent_Name': 'Continent'}, inplace=True)
```

```python
df.groupby('Continent')['Unemployment By %'].sum().plot(kind='bar', ylabel='Unemployment By %', title='Unemployment By % per Continent', color='red')
```

```python
df.groupby('Continent')['Unemployment By %'].sum().plot(kind='line', ylabel='Unemployment By %', title='Unemployment By % per Continent', color='blue')
```

```python
df.groupby('Continent')['Unemployment By %'].sum().plot(kind='pie', ylabel='Unemployment By %', title='Unemployment By % per Continent', autopct='%1.1f%%')
```

```python
plt.plot(df['Continent'], df['Unemployment By %'], 'o', color='black')
plt.xlabel('Continent')
plt.ylabel('Unemployment By %')
plt.title('Unemployment By % per Continent')
plt
```

```python
df.groupby('Continent')['Unemployment By %'].mean().plot(kind='pie', ylabel='Unemployment By %', title='Unemployment By % per Continent', autopct='%1.1f%%', figsize=(10,10))
```
