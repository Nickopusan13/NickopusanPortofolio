```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\Air bnb\Airbnb Raw.csv", low_memory=False)
df
```

```python
df.shape
```

```python
df.sort_values(by="NAME")
```

```python
df.drop(columns=["license", "last review", "reviews per month"], inplace=True)
```

```python
df['house_rules'].fillna('No rules given', inplace=True)
```

```python
df = df.drop_duplicates()
```

```python
df.dropna(subset='NAME', inplace=True)
```

```python
df.dropna(subset=['country', 'country code'], how='all', inplace=True)
```

```python
df.shape
```

```python
df[df['country'].isna()]
```

```python
df.loc[df['country code'] == 'US', 'country'] = df.loc[df['country code'] == 'US', 'country'].fillna('United States')
df.loc[df['country'] == 'United States', 'country code'] = df.loc[df['country'] == 'United States', 'country code'].fillna('US')
```

```python
df['host_identity_verified'].fillna('unconfirmed', inplace=True)
df['host name'].fillna('Unknown', inplace=True)
df['neighbourhood'].fillna('Unknown', inplace=True)
df['neighbourhood group'].fillna('Unknown', inplace=True)
```

```python
# Remove dollar signs and convert to integers
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype('Int64')
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype('Int64')

df['Construction year'] = df['Construction year'].astype('Int64')
df['minimum nights'] = df['minimum nights'].astype('Int64')
df['number of reviews'] = df['number of reviews'].astype('Int64')
df['review rate number'] = df['review rate number'].astype('Int64')
df['calculated host listings count'] = df['calculated host listings count'].astype('Int64')
df['availability 365'] = df['availability 365'].astype('Int64')
```

```python
df.head()
```

```python
df.rename(columns={'price': 'Price $', 'service fee': 'Service Fee $'}, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='Price $', y='Service Fee $')
plt.title('Relationship between Price and Service Fee')
plt.xlabel('Price ($)')
plt.ylabel('Service Fee ($)')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(data=df, x='availability 365', y='room type', ci=None)
plt.title('Availability of Listings Throughout the Year by Room Type')
plt.xlabel('Availability (days)')
plt.ylabel('Room Type')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['review rate number'], bins=20, kde=True)
plt.title('Distribution of Review Ratings')
plt.xlabel('Review Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Construction year'].dropna(), bins=30, kde=True)
plt.title('Distribution of Listings by Construction Year')
plt.xlabel('Construction Year')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
room_type_counts = df['room type'].value_counts()
explode = [0.1 if i == room_type_counts.idxmin() else 0 for i in room_type_counts.index]
room_type_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('pastel'),
    pctdistance=0.85,
    explode=explode,
    textprops={'fontsize': 12}
)
plt.title('Distribution of Room Types', fontsize=14)
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='instant_bookable')
plt.title('Distribution of Instant Bookable Listings')
plt.xlabel('Instant Bookable')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
df['host_identity_verified'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Host Identity Verification Status')
plt.ylabel('')
plt.show()
```
