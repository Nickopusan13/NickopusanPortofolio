```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

print(plt.style.available)
plt.style.use('grayscale')
```

```python
df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\Student Performance\Student Performance Factors.csv")
df.head()
```

```python
print(f"""In this data we have {df.shape[0]} rows and {df.shape[1]} columns
and we have {df.duplicated().sum()} duplicate with null data:
{df.isnull().sum()}""")
```

```python
df = df.dropna()
```

```python
# Pie Chart
gender_counts = df['Gender'].value_counts(normalize=True) * 100
plt.figure(figsize=(5, 5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['blue','pink'])
plt.title('Gender Percentage Distribution')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Exam_Score'], bins=30, kde=True)
plt.title("Distribution of Exam Scores")
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot to visualize the relationship between teacher quality and exam scores
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Teacher_Quality', y='Exam_Score', data=df)
plt.title('Relationship between Teacher Quality and Exam Scores')
plt.xlabel('Teacher Quality')
plt.ylabel('Exam Score')
plt.show()
```

```python
plt.Figure(figsize=(10, 5))
sns.boxplot(x=df['Physical_Activity'], y=df['Exam_Score'])
plt.title("Effect of Physical Activity on Exam Score")
plt.xlabel("Physical Activity")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.Figure(figsize=(10,5))
sns.violinplot(x='Learning_Disabilities', y='Exam_Score', data=df)
plt.title("Exam Score Distribution by Learning Disabilities")
plt.xlabel("Learning Disabilities (Yes/No)")
plt.ylabel("Exam Score")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['Parental_Education_Level'], y=df['Exam_Score'], estimator=np.mean)
plt.title('Average Exam Score by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Average Exam Score')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Distance_from_Home'], df['Exam_Score'], alpha=0.5)
plt.title('Distance from Home vs. Exam Score')
plt.xlabel('Distance from Home (km)')
plt.ylabel('Exam Score')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Gender'], y=df['Exam_Score'])
plt.title('Exam Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Exam Score')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
