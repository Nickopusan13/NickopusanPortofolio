```python
import pandas as pd
```

```python
# Import the data
t_rating = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\tourism_rating.csv")
t_id = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\tourism_id.csv")
t_package = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\package_tourism.csv")
t_user = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\user.csv")
```

```python
# Table Rating
t_rating
```

```python
# Table ID
t_id
```

```python
# Table Package
t_package
```

```python
# Table User
t_user
```

```python
# Merge Table
df = pd.merge(t_user, t_rating, on='User_Id')
df
```

```python
# Merge Table
df = pd.merge(df, t_id, on='Place_Id')
df
```

```python
# Drop the unnecessary table
df.drop(['Coordinate', 'Lat', 'Long', 'Time_Minutes'], axis=1, inplace=True)
```

```python
# See duplicate
df.duplicated().sum()
```

```python
# Delete duplicate
df.drop_duplicates()
```

```python
# Make column province
df['Province'] = df['Location'].apply(lambda x: x.split(', ')[-1])
df.head()
```

```python
# Delete province in location
df['Location'] = df['Location'].apply(lambda x: x.split(',')[0])
df.head()
```

```python
# Delete city
df.drop('City', axis=1, inplace=True)
```

```python
# Save to csv
df.to_csv("Tourism Clean.csv", index=False)
```
