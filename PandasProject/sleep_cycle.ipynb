```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\sleep_cycle_productivity.csv")
```

```python
print(f"""This data have {df.shape[0]} column and {df.shape[1]} rows and have {df.duplicated().sum()} duplicate. This table have null :
{df.isnull().sum()}""")
```

```python
plt.figure(figsize=(10,6))
# Distributin of total sleep hours
plt.subplot(1,2,1)
sns.histplot(df['Total Sleep Hours'], bins=20, kde=True, color='blue')
plt.title("Distributin of Total Sleep Hours")
plt.xlabel("Total Sleep Hours")
plt.ylabel("Frequency")
# Distributin of sleep quality
plt.subplot(1,2,2)
sns.histplot(df['Sleep Quality'], bins=20, kde=True, color='pink')
plt.title("Distributin of Sleep Quality")
plt.xlabel("Sleep Quality (1-10)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
plt.figure(figsize=(16, 6))
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Sleep Quality', hue='Gender', alpha=0.6)
plt.title('Age vs Sleep Quality')
plt.xlabel('Age')
plt.ylabel('Sleep Quality')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Gender', y='Sleep Quality', palette='Set3')
plt.title('Sleep Quality by Gender')
plt.xlabel('Gender')
plt.ylabel('Sleep Quality')
plt.show()
```
