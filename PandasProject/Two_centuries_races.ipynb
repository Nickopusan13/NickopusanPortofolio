```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Masterpiece By Nickopusan\Document\Two Centuries Races\TWO_CENTURIES_OF_UM_RACES.csv", low_memory=False)
```

```python
df.info()
```

```python
df['Athlete year of birth'] = df['Athlete year of birth'].astype('Int64')
```

```python
df['Athlete average speed'] = pd.to_numeric(df['Athlete average speed'], errors='coerce')
df['Event dates'] = pd.to_datetime(df['Event dates'], errors='coerce')
df['Event dates'] = df['Event dates'].dt.strftime('%Y-%m-%d')
df = df.astype({
    "Athlete average speed": "float64",
})
```

```python
df = df.dropna(subset=['Athlete average speed', 'Athlete year of birth', 'Athlete age category'], how="all")
```
